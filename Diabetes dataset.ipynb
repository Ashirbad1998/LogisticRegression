### Feature Engineering

#### Converting Categorical variable to Numerical Variable

Apply 'get_dummies()' Method

This is one of the approach and also an each one to encode Categorical data. pandas get_dummies() method takes categorical feature as an argument. Then it creates a Dummy variable for every label in the feature, such that each dummy variable holds data as 1 or 0.1 indicates the presence of a particular label and 0 indicates the absence of a particular label.

```python
df1 = pd.get_dummies(new_df,drop_first=True)
```

```python
df1.shape
```

### Split the dataframe into x & y

```python
# Separate object for target feature 
y = df1["diabetes"]

# Separate object for Predictors 
X = df1.drop("diabetes",axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
print('X Shape =',X.shape,'\ny Shape =',y.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve,roc_auc_score
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
LR=LogisticRegression()
LR.fit(X_train,y_train)
```

```python
train_pred = LR.predict(X_train)
test_pred = LR.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_train,train_pred))
```

```python
print(confusion_matrix(y_test,test_pred))
```

```python
print(classification_report(y_train,train_pred))
```

```python
print(classification_report(y_test,test_pred))
```

### Roc Curve

```python
fpr,tpr,threshold=roc_curve(y_test,test_pred)
```

```python
print('roc_auc_score for logistic Regression= ',roc_auc_score(y_test,test_pred))
```

```python
from sklearn.metrics import auc
auc=auc(fpr,tpr)
```

```python
plt.subplots(1,figsize=(10,10))
plt.title('Receiver Operating Characteristic - Logistic regression')
plt.plot(fpr,tpr,label='auc= {}'.format(auc))
plt.legend(loc='lower right')
plt.plot([0, 1], ls="--")
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```
